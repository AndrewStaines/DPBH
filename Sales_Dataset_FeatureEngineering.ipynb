```python
df.rating_count.replace(np.nan,"9378",inplace=True)
```

```python
df.info()
```

```python
for i in range (0,1465):
  x=df['discounted_price'][i]
  df['discounted_price'][i]=x[1::]
```

```python
for i in range (0,1465):
  x=df['discounted_price'][i]
  df['discounted_price'][i]=''
  for j in x:
    if(j!=','):
      df['discounted_price'][i]+=j
    else:
      continue;
```

```python
for i in range (0,1465):
  x=df['actual_price'][i]
  df['actual_price'][i]=x[1::]
```

```python
for i in range (0,1465):
  x=df['actual_price'][i]
  df['actual_price'][i]=''
  for j in x:
    if(j!=','):
      df['actual_price'][i]+=j
    else:
      continue;
```

```python
for i in range (0,1465):
  x=df['discount_percentage'][i]
  if(len(x)==3):
    df['discount_percentage'][i]=x[:2:]
  else:
    df['discount_percentage'][i]=x[:1:]
```

```python
for i in range (0,1465):
  x=df['rating_count'][i]
  df['rating_count'][i]=''
  for j in x:
    if(j!=','):
      df['rating_count'][i]+=j
    else:
      continue;
```
