```python
for i in range (0,1465):
  x=df['discounted_price'][i]
  df['discounted_price'][i]=x[1::]
```

```python
for i in range (0,1465):
  x=df['discounted_price'][i]
  df['discounted_price'][i]=''
  for j in x:
    if(j!=','):
      df['discounted_price'][i]+=j
    elif(j=='.'):
      break
    else:
      continue;
```

```python
for i in range (0,1465):
  x=df['actual_price'][i]
  df['actual_price'][i]=x[1::]
```

```python
for i in range (0,1465):
  x=df['actual_price'][i]
  df['actual_price'][i]=''
  for j in x:
    if(j!=','):
      df['actual_price'][i]+=j
    elif(j=='.'):
      break
    else:
      continue;
```

```python
for i in range (0,1465):
  x=df['discount_percentage'][i]
  if(len(x)==3):
    df['discount_percentage'][i]=x[:2:]
  else:
    df['discount_percentage'][i]=x[:1:]
```

```python
for i in range (0,1465):
  x=df['rating_count'][i]
  df['rating_count'][i]=''
  for j in x:
    if(j!=','):
      df['rating_count'][i]+=j
    else:
      continue;
```

```python
df[['discount_percentage','rating_count']]=df[['discount_percentage','rating_count']].astype(int)
```

```python
df[['discounted_price','actual_price']]=df[['discounted_price','actual_price']].astype(float)
```
